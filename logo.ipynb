```
import numpy as np
import imcmc
```

```
seed = 0
```

```
# Rerun this cell until the image looks good
seed += 1
np.random.seed(10)
print(seed)
image = imcmc.load_image('iris.png', 'L')
trace = imcmc.sample_grayscale(image, samples=1000, tune=500, nchains=10)
imcmc.plot_multitrace(trace, image, marker='o', linestyle='', alpha=0.5, colors=['#0000FF']);
```

```
# Then run this step on the final image to get the samplers drawing it
imcmc.make_gif(trace, image, dpi=40, marker='o', 
               alpha=0.5, colors=['#0000FF'], filename='iris.gif')
```
